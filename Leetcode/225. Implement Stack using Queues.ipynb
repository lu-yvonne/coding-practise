```python
# Implement the following operations of a stack using queues.

# push(x) -- Push element x onto stack.
# pop() -- Removes the element on top of the stack.
# top() -- Get the top element.
# empty() -- Return whether the stack is empty.

# Example:
# MyStack stack = new MyStack();
# stack.push(1);
# stack.push(2);  
# stack.top();   // returns 2
# stack.pop();   // returns 2
# stack.empty(); // returns false
```

Notes:
<br>You must use only standard operations of a queue -- which means only push to back, peek/pop from front, size, and is empty operations are valid.
<br>Depending on your language, queue may not be supported natively. You may simulate a queue by using a list or deque (double-ended queue), as long as you use only standard operations of a queue.
<br>You may assume that all operations are valid (for example, no pop or top operations will be called on an empty stack).

```python
# 思路
# 堆：后进先出
# 队列：先进先出

# 我们可以用两个队列来实现，一个作为主队列，另一个作为辅助队列，

# 例子：1 -> 2 ->3                      stack :  3
# 首先依次压进队列queue1：                       2
#queue1:      1                                  1
#             2 
#             3 （从上到下） 
# 在python中为[3,2,1]

# 依次弹出并压进队列queue2，python中的pop()函数是弹出list中的最后一个数
#queue2:      3
#             2
#             1 （从上到下） 
# 在python中为[1,2,3]

# 注意在python中append实际上是栈的实现方式，因为一次append1,2,3，最后的输出为[1,2,3]，top是3而不是1
# 所以每次入栈的时候都插入索引值为0的地方

# 其实这个没什么技术含量，因为在python里pop和push的实现模式其实都是stack
# 直接用一个list也可以完成，但是不符合题意

class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.q1 = [] # 主队列
        self.q2 = [] # 辅助队列

    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        self.q1.insert(0,x)
        

    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns that element.
        """
        if not self.q1:
            return None
        else:
#             构造一个相同的除了进q1的第一个元素，在python的list中是最后一个元素，其他与q1相同的q1
            while len(self.q1) != 1:
                self.q2.insert(0,self.q1.pop())
#             第一个元素为q1剩下没有被pop的元素
            top = self.q1.pop()
#             此时q1为空
#             交换q1和q2,将q2重置为空,q1为pop一个先进元素后的队列
            self.q1, self.q2 = self.q2, self.q1 # 同
            return top
        
    def top(self) -> int:
        """
        Get the top element.
        """
        if not self.q1:
            return None
        else:
            for i in range(len(self.q1)):
                self.q2.insert(0,self.q1[i])
            top = self.q2[-1]
            self.q2 = []
            return top
#         这个部分可能写复杂了？？？
        
    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        if len(self.q1) ==0:
            return True
        else:
            return False

# TEST
# Your MyStack object will be instantiated and called as such:
obj = MyStack()
obj.push(1)
obj.push(2)
obj.push(3)
# param_2 = obj.top()
param_2 = obj.pop()
print(obj.q1)
print(obj.q2)
param_3 = obj.top()
param_4 = obj.empty()
print(param_2)
print(param_3)
print(param_4)
```
