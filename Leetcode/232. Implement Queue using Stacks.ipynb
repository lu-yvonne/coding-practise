Implement the following operations of a queue using stacks.
<br>push(x) -- Push element x to the back of queue.
<br>pop() -- Removes the element from in front of queue.
<br>peek() -- Get the front element.
<br>empty() -- Return whether the queue is empty.

Example:
<br>MyQueue queue = new MyQueue();
<br>queue.push(1);
<br>queue.push(2);
<br>queue.peek();  // returns 1
<br>queue.pop();   // returns 1
<br>queue.empty(); // returns false

Notes:
<br>You must use only standard operations of a stack -- which means only push to top, peek/pop from top, size, and is empty operations are valid.
<br>Depending on your language, stack may not be supported natively. You may simulate a stack by using a list or deque (double-ended queue), as long as you use only standard operations of a stack.
<br>You may assume that all operations are valid (for example, no pop or peek operations will be called on an empty queue).

```python
# 思路

# 首先明确栈和队列的特点：
# 栈：后进先出
# 队列：先进先出

# 所以我们可以用两个栈来实现，一个作为输入栈，另一个作为输出栈，
# 如果需要使用push和peek的功能，则查找输出栈内有没有输入栈的内容，如果有，则使用输出栈的功能
# 如果没有，则把所有输入栈内的元素放进输出栈中去，再使用输出栈的功能
# 例子：1 -> 2 ->3 
# 首先依次压进栈stack1
#stack1:      3 
#             2 
#             1 （从上到下） 
# 在python中为[1,2,3]
# 依次弹出并压进栈stack2，python中的pop()函数是弹出list中的最后一个数
# stack2:     1 
#             2 
#             3  (从上到下) 
# 在python中为[3,2,1]
# 然后依次弹出：1 2 3或查看栈顶元素：1，则达到队列效果


class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.s1 = []
        self.s2 = []

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.s1.append(x)

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        if self.s2:
            return self.s2.pop()
        elif not self.s1:
            return None
        else:
            while self.s1:
                self.s2.append(self.s1.pop())
            return self.s2.pop()

    def peek(self) -> int:
        """
        Get the front element.
        """
        if self.s2:
            return self.s2[-1]
        elif not self.s1:
            return None
        else:
            while self.s1:
                self.s2.append(self.s1.pop())
            return self.s2[-1]

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        if len(self.s1)==0 and len(self.s2)==0:
            return True
        else:
            return False

# Test
# Your MyQueue object will be instantiated and called as such:
obj = MyQueue()
obj.push(1)
obj.push(2)
# print(obj.s1)
param_2 = obj.peek()
# print(obj.s2)
param_3 = obj.pop()
param_4 = obj.empty()
print(param_2)
print(param_3)
print(param_4)
```
