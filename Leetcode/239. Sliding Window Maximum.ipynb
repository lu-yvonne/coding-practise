```python
# Given an array nums, there is a sliding window of size k which is moving from the very left of the array to the very right.
# You can only see the k numbers in the window. Each time the sliding window moves right by one position.
# Return the max sliding window.

# Example:
# Input: nums = [1,3,-1,-3,5,3,6,7], and k = 3
# Output: [3,3,5,5,6,7] 
# Explanation: 

# Window position                Max
# ---------------               -----
# [1  3  -1] -3  5  3  6  7       3
#  1 [3  -1  -3] 5  3  6  7       3
#  1  3 [-1  -3  5] 3  6  7       5
#  1  3  -1 [-3  5  3] 6  7       5
#  1  3  -1  -3 [5  3  6] 7       6
#  1  3  -1  -3  5 [3  6  7]      7
# Note: 
# You may assume k is always valid, 1 ≤ k ≤ input array's size for non-empty array.

# Follow up:
# Could you solve it in linear time?
```

```python
# 思路
# 方法1
# 利用优先队列，构造大顶堆，堆顶元素为最大元素
# 1.维护Heap，删除离开的元素，加入新的元素，保持heap的元素个数为K
# 2.Max -> Top
# 依次进行下去即可
# 时间复杂度：N*logK*O(1) = NlogK

import heapq
class Solution:
    def maxSlidingWindow(self, nums, k: int):
        if not nums:
            return None
        self.data = []
        self.k = k
        self.res = []
#         if len(nums) <= k:
#             return max(nums)

        for i in range(0,self.k):
            self.data.append(-nums[i])
            heapq.heapify(self.data)
#         构造初始堆，size = K
        self.res.append(-self.data[0])
#         输出初始堆的最大值
        for i in range(k,len(nums)):
#             print(self.data)
#             print(nums[i-k])
#             print(nums[i])
            if nums[i-k] == -self.data[0]: # 如果需要删除的数是堆的最大值，直接pop，然后再push新的数进堆
                heapq.heappop(self.data)
                self.Heap(nums[i])
#                 print(self.data)
            else: # 不然直接将新的数与堆顶作比较，比堆顶大则替换，小则加入堆，删掉需要删掉的数并且重新排序
                tmp = self.data[0]
                self.Heap(nums[i])
#                 print(self.data)
                if nums[i]!= -self.data[0]:
                    self.data.append(-nums[i])
#                     print(self.data)
#                     print(nums[i-k])
                    self.data.remove(-nums[i-k])
                    self.data = sorted(self.data) # 可能就是这一步造成了时间复杂度更高 需要改进
                else: 
                # 如果新加进的值是最大值，而覆盖了前一个数组的最大值，需要将前一个数组最大值补上，
                # 并且删除需要删除的部分
                    self.data.remove(-nums[i-k])
                    self.data.append(tmp)
                    
#             print(self.data)
#             print(nums[i-k])
#             print(nums[i-k])
        return self.res
        
#     建立最大堆
    def Heap(self,val):
        val = -val
        if len(self.data) < self.k:
            heapq.heappush(self.data, val)
            self.res.append(-self.data[0])
        elif val < self.data[0]:
            heapq.heapreplace(self.data,val)
#            输出最大值，将最大值加进result数组
            self.res.append(-self.data[0])
        else:
            self.res.append(-self.data[0])
        
obj = Solution()
test1  = obj.maxSlidingWindow([-95,92,-85,59,-59,-14,88,-39,2,92,94,79,78,-58,37,48,63,-91,91,74,-28,39,90,-9,-72,-88,-72,93,38,14,-83,-2,21,4,-75,
                               -65,3,63,100,59,-48,43,35,-49,48,-36,-64,-13,-7,-29,87,34,
                               56,-39,-5,-27,-28,10,-57,100,-43,-98,19,-59,78,-28,-91,67,41,-64,76,5,-58,-89,83,
                               26,-7,-82,-32,-76,86,52,-6,
                               84,20,51,-86,26,
                               46,35,-23,30,-51,54,
                               19,30,27,80,
                               45,22],10)
# test1
```
